# Actual Function

```python
def channel_swap(f_name, in_dir=raw, out_dir=proc):
    fin = path.join(in_dir, f_name)
    fout = path.join(out_dir, f_name)
    data = imread(fin)
    swap = data[:, :, [2, 1, 0]]
    p_img = Image.fromarray(swap, 'RGB')
    p_img.save(fout)
```

```python
channel_swap('IMG_0070.JPG')
```

```python
img_list = [i for i in listdir(raw) if i.split('.')[-1].lower() == 'jpg']
img_list
```

```python
for f in img_list:
    channel_swap(f)
```
